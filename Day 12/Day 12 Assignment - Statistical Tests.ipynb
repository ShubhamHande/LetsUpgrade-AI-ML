```python
a_yes=dataset[dataset['Attrition']=='Yes']
a_no=dataset[dataset['Attrition']=='No']
```

```python
from scipy.stats import ttest_ind
stats,p=ttest_ind(a_yes.Age, a_no.Age)
print(stats,p)
```

__Hypothesis 1 Age v/s Attrition__

__TTest__

H0(Null Hypothesis):- There is no significance difference between the Age with respect to Attrition

Ha(Alternative Hypothesis):- There is a significance difference between the Age with respect to Attrition

p value :- 1.9968016158894556e-26 , which is less than 0.05

__Reject H0__

__Conclusion :- There is a significance difference between the Age with respect to Attrition__

```python
from scipy.stats import mannwhitneyu
stats,p=mannwhitneyu(a_yes.MonthlyIncome, a_no.MonthlyIncome)
print(stats,p)
```

__Hypothesis 2 Monthly Income v/s Income__

__MannwhitneyTest__

H0(Null Hypothesis):- There is no significance difference between the Age with respect to MonthlyIncome

Ha(Alternative Hypothesis):- There is a significance difference between the Age with respect to MonthlyIncome

p value :- 0.053577283839938566 , which is greater than 0.05

__Accept H0__

__Conclusion :- There is no significance difference between the Age with respect to MonthlyIncome__
