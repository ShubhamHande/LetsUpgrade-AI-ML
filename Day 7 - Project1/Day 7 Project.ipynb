```python
fig, ax = plt.subplots()
my_scatter_plot = ax.scatter(dataset["YearsAtCompany"],dataset["MonthlyIncome"])
ax.set_xlabel("Employee working Years in Company")
ax.set_ylabel("Monthly Income Given to Employee")

ax.set_title("YearsAtCompany v/s Monthly Income Given")
plt.show()
```

As we Can See on above Scatterplot, Employees who had worked for more years are getting less salary and Who have worked for 0 or less than 5 years are getting more salary
Of course other factors like Education, Total no. of Years are also matter

```python
fig, ax = plt.subplots()
my_scatter_plot = ax.scatter(dataset["YearsAtCompany"],dataset["PercentSalaryHike"])
ax.set_xlabel("Employee working Years in Company")
ax.set_ylabel("Percent salary Hike received by Employee")

ax.set_title("YearsAtCompany v/s Percent Salary Hike")
plt.show()
```
